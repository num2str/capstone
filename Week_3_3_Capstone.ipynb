```python
d1 = pd.read_csv("Geospatial_Coordinates.csv")
```

```python
d1.head()
```

```python
Webpage = pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M')
d2 = Webpage[0]

d2['Borough'].replace("Not assigned", np.nan, inplace=True)
d2.dropna(subset=['Borough'], axis=0, inplace=True)
d2.reset_index(drop=True, inplace=True)
d2
```

```python
# d3.drop(['Postal Code'],axis=1)

d4 = pd.merge(d2, d1, on='Postal Code', how='inner')
```

Create Map

```python
# create map of Toronto using latitude and longitude values 

map_clusters = folium.Map(location=[43.753259, -79.329656], zoom_start=10.7)

 

# add markers to map 

for lat, lng, borough, neighbourhood in zip(d4['Latitude'], d4['Longitude'],d4['Borough'], d4['Neighbourhood']): 

     

    label = '{}, {}'.format(neighbourhood, borough) 

    label = folium.Popup(label, parse_html=True) 

    folium.CircleMarker( 

        [lat, lng], 

        radius=5, 

        popup=label, 

        color='blue', 

        fill=True, 

        fill_color='#3186cc', 

        fill_opacity=0.7).add_to(map_clusters)   

     

map_clusters

```
